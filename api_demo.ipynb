```python
import requests
from codecs import decode
import pandas as pd
import json
```

```python
url = 'http://127.0.0.1:5000'
ret = requests.get(url)

print(ret.url)
print(ret.text)
```

```python
payload = {'idiom': '坚定不移', 'length': 2}
ret = requests.get("http://127.0.0.1:5000/find_next", params=payload)
content = decode(ret.text, 'unicode_escape')

# print unicode
# print(u'\u575a\u5b9a\u4e0d\u79fb')

print(ret.url)
print(content)
```

```python
pd.DataFrame(json.loads(content))
```
